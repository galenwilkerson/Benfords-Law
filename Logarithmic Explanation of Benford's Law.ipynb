# Benford's Law Visualization in Logarithmic Space

Benford's Law states that in many naturally occurring collections of numbers, the leading digit \(d\) occurs with probability
$$
P(d) = \log_{10}\left(1 + \frac{1}{d}\right)
$$

This can be visualized on a number line using the logarithmic scale. For a number to have a leading digit \(d\), its logarithm (base 10) must lie within the interval:
$$
\log_{10}(d) \leq \log_{10}(x) < \log_{10}(d+1)
$$

```python
# Let's plot these intervals on a number line:

import matplotlib.pyplot as plt
import numpy as np

# Logarithmic intervals for leading digits
intervals = {
    1: (0, np.log10(2)),
    2: (np.log10(2), np.log10(3)),
    3: (np.log10(3), np.log10(4)),
    4: (np.log10(4), np.log10(5)),
    5: (np.log10(5), np.log10(6)),
    6: (np.log10(6), np.log10(7)),
    7: (np.log10(7), np.log10(8)),
    8: (np.log10(8), np.log10(9)),
    9: (np.log10(9), np.log10(10))
}

# Plotting the number line
fig, ax = plt.subplots(figsize=(10, 2))

# Draw number line
ax.plot([0, 1], [0, 0], 'k', lw=2)
ax.set_xticks(np.linspace(0, 1, 11))
ax.set_xticklabels(np.round(np.linspace(0, 1, 11), 2))
ax.set_yticks([])
ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.2, 0.2)
ax.set_xlabel('Log values')

# Highlight intervals
colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown', 'cyan', 'magenta', 'gray']
for i, (digit, (start, end)) in enumerate(intervals.items()):
    ax.plot([start, end], [0.1, 0.1], color=colors[i], lw=8, label=f'Leading digit {digit}')

# Annotate the intervals
for digit, (start, end) in intervals.items():
    ax.text((start + end) / 2, 0.15, f'{digit}', horizontalalignment='center', fontsize=12, color='black')

# Move legend outside the figure
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Benford's Law: Leading Digit Intervals in Logarithmic Space")
plt.grid(False)
plt.show()
```

# Benford's Law is really due to the shape of this curve.  Notice that a wider range of values start with 1 (between 0 and 20), versus 2, 3, etc.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 1 to 100 to avoid log10(0) issue
x = np.arange(1, 10, 0.1)

# Plot the logarithmic curve with colored horizontal regions for each leading digit
plt.figure(figsize=(10, 6))

# Define colors for each leading digit
colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown', 'cyan', 'magenta', 'gray']

# Highlight intervals for leading digits
for digit in range(1, 10):
    start = digit
    end = start + 1
    plt.fill_between(x, np.log10(start), np.log10(end), color=colors[digit - 1], alpha=0.2, label=f'Starts with {digit}')

# Plot the logarithmic curve
plt.plot(x, np.log10(x), label='log10(x)', color='black')
plt.xlabel('x')
plt.ylabel('log10(x)')
plt.title("Benford's Law is really due to the shape of this curve. Notice that a wider range of values start with 1 (between 1 and 10), versus 2, 3, etc.")

# Move legend outside the figure
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(False)
plt.xlim([0, 10])
plt.ylim([0, 1])
plt.show()
```
